```python
from scipy.sparse import load_npz
import numpy as np

P = load_npz("/Users/sam/Documents/Wesleyan/Thayer_Lab/2025/summer/bharat/sparse_matrix/P_transition.npz")

absorbing = []
for i in range(P.shape[0]):
    row = P.getrow(i).toarray().ravel()
    is_absorbing = np.isclose(row[i], 1.0) and np.isclose(row.sum(), 1.0) and np.count_nonzero(row) == 1
    if is_absorbing:
        absorbing.append(i)

print(f"# Absorbing nodes: {len(absorbing)}")
print("Absorbing node indices:", absorbing)
```

```python
for i in absorbing:
    out_degree = P.getrow(i).count_nonzero()
    print(f"Node {i} row sum: {P.getrow(i).sum()}, nonzero: {out_degree}, diagonal: {P[i, i]}")
```

```python
import networkx as nx
from scipy.sparse import load_npz
from networkx import DiGraph
```

```python
# Load your matrix
P = load_npz("/Users/sam/Documents/Wesleyan/Thayer_Lab/2025/summer/bharat/sparse_matrix/P_transition.npz")  # transition matrix

# Convert to a directed graph (weighted by transition probabilities)
G = nx.from_scipy_sparse_array(P, create_using=DiGraph)
```

```python
is_absorbing = (P != 0).sum(axis=1).A1 == 1  # as before

for i in G.nodes:
    G.nodes[i]['state_type'] = 'absorbing' if is_absorbing[i] else 'transient'
```

```python
# Only absorbing if there's a single edge and it is a self-loop with prob 1
absorbing_nodes = []
for i in range(P.shape[0]):
    row = P.getrow(i).toarray().ravel()
    if np.isclose(row[i], 1.0) and np.count_nonzero(row) == 1:
        G.nodes[i]['state_type'] = 'absorbing'
        absorbing_nodes.append(i)
    else:
        G.nodes[i]['state_type'] = 'transient'

print(f"Absorbing nodes marked in graph: {absorbing_nodes}")
```

```python
nx.write_graphml(G, "transition_network.graphml")
```
